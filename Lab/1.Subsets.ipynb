```
from pip._vendor.distlib.compat import raw_input 
#given the sample array
SampleArray = []
#taking the input from the user
elem01 = int(raw_input("How many element can be added:"))
#looping for the getting the values to the array
for p in range(0, elem01):
    SampleArray.append(int(raw_input("Enter next no :")))
m = len(SampleArray)
#Declaring the empty list 
_list = []
#using for loop for the range of 2 power m
for p in range(2**m):
    simpStr = ""
    for a in range(m):
       if (p & (1 << a)) != 0:
        simpStr += str(SampleArray[a]) + "|"     
    if simpStr not in _list and len(simpStr) > 0:
       _list.append(simpStr)      
for simpStr in _list:
    SampleArray = simpStr.split('|')
    for string in SampleArray:
        print(string, end = " ")
    print()
```
