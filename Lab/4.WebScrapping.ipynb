```
from bs4 import BeautifulSoup
import urllib.request
import os
url = "https://catalog.umkc.edu/course-offerings/graduate/comp-sci/"
source_code = urllib.request.urlopen(url)
plain_text = source_code
soup = BeautifulSoup(plain_text, "html.parser")
result_tags = soup.find_all('span', {"class": "title"})
result_content = soup.find_all('p', {"class": "courseblockdesc"})
with open("hello.txt", "w") as file:
  for result in result_tags:
    file.write(result.text + '\n')
  for content in result_content:
    file.write(content.text + '\n')
```
