```
noOfdictt = int(input("Entr how many elements needed in dictionary_1:"))
dictt_1 = {}

for i in range(noOfdictt):
    keys = input("Enter the Key : ") # here i have taken keys as strings
    values = int(input("Enter the value : ")) # here i have taken values as integers

    dictt_1[keys] = values
print("User input for the dictionary_1:", dictt_1)

print(" ")
noOfdictt2 = int(input("Enter how many elements needed in dictionary_2:"))
dictt_2 = {}

for i in range(noOfdictt2):
    keys = input("Enter the Key : ") # here i have taken keys as strings
    values = int(input("Enter the value : ")) # here i have taken values as integers
    dictt_2[keys] = values
print("User input for the dictionary_2:", dictt_2)

#Concatnate
data_1 = dict(dictt_1)
data_1.update(dictt_2)
print(" ")
print("Unsorted_dictonary:", data_1)
b1 = sorted(data_1.items(), key=lambda x: x[1]) 
# key value pairs are sorted using lambda and sorted functions
print(" ")
print("Sorted_dictonary:", b1)
```
